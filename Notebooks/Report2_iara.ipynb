### Derived Metrics - Iara

**Percentage of Installations by Voltage Level per District**

According to E-redes ([Manual de Ligações à Rede](https://provedordocliente.e-redes.pt/Files/PDF/Manual-de-Ligacoes-a-Rede.pdf)):

*High Voltage (AT)*
- Companies with capacities >10 MVA, supplied at 60 kV.
- Clear proxy for heavy industry or large commercial facilities.
- Districts with a higher proportion of AT installations → more industrialized.

*Medium Voltage (MT)*
- Companies with capacities >200 kVA, voltages of 10–30 kV.
- Also indicates industrial activity or large commercial companies.
- Complements AT; districts with a higher proportion of MT → more industrialized areas, but less intense than AT.

*Low Voltage Normal (BTN) and Low Voltage Special (BTE)*
- **BTN** → residences, small shops/offices.
- **BTE** → small/medium companies (>41.4 kVA).
- Districts dominated by BTN → mainly residential areas.
- BTE is mixed, can indicate areas with small industries or commerce, but less significant than MT/AT.


```python
# Make a copy
df = df_filtered.copy()

# Group by Quarter, District, and Voltage level, summing the Number of installations
grouped = (
    df.groupby(['Quarter', 'District', 'Voltage level'], as_index=False)['Number of installations'].sum()
)

# Pivot table to have Voltage levels as columns
pivot = grouped.pivot_table(
    index=['Quarter', 'District'],
    columns='Voltage level',
    values='Number of installations',
    fill_value=0
).reset_index()

# Ensure all expected voltage columns exist
for col in ['AT', 'MT', 'BTN', 'BTE']:
    if col not in pivot.columns:
        pivot[col] = 0

# Calculate total installations per row
pivot['Total'] = pivot[['AT','MT','BTN','BTE']].sum(axis=1)

# Calculate percentage per voltage level
pivot["District_High_Voltage_AT(%)"] = pivot['AT'] / pivot['Total'] * 100
pivot["District_Medium_Voltage_MT(%)"] = pivot['MT'] / pivot['Total'] * 100
pivot["District_Low_Voltage_BTN(%)"] = pivot['BTN'] / pivot['Total'] * 100
pivot["District_Low_Voltage_BTE(%)"] = pivot['BTE'] / pivot['Total'] * 100

# Select only the relevant columns
df_result = pivot[['Quarter', 'District',
                   'District_High_Voltage_AT(%)',
                   'District_Medium_Voltage_MT(%)',
                   'District_Low_Voltage_BTN(%)',
                   'District_Low_Voltage_BTE(%)']]

df_result
```

```python
df_final = df_filtered.merge(df_result, on=['Quarter', 'District'], how='left')
df_final.head(10)
```

### RQ2: Compare the evolution of installed capacity between 2023 and 2024 across residential and industrial UPACs to assess differences in growth patterns. - Iara and Mariana

```python
df=df_final
```

```python
df["Sector_Type"] = df["Voltage level"].apply(
    lambda x: "Industrial" if x in ["MT", "AT"] else "Residential"
)

df["Sector_Type"] = np.where(df["Voltage level"].isin(["MT", "AT"]),
                             "Industrial", "Residential")
```

**Shapiro-Wilk test de normalidade**

```python
df['Year'] = df['Quarter'].astype(str).str.extract(r'(20\d{2})').astype(int)

from scipy.stats import shapiro

grupos = df.groupby(['Year', 'Sector_Type'])

for (ano, setor), grupo in grupos:
    stat, p = shapiro(grupo['Total installed power (kW)'])
    print(f"{ano} - {setor}: W={stat:.3f}, p={p:.3f}")
```

**Teste de Kruskal-Wallis**

```python
g1 = df[(df['Year']==2023) & (df['Sector_Type']=='Residential')]['Total installed power (kW)']
g2 = df[(df['Year']==2024) & (df['Sector_Type']=='Residential')]['Total installed power (kW)']
g3 = df[(df['Year']==2023) & (df['Sector_Type']=='Industrial')]['Total installed power (kW)']
g4 = df[(df['Year']==2024) & (df['Sector_Type']=='Industrial')]['Total installed power (kW)']


stat, p = kruskal(g1, g2, g3, g4)
print(f'Kruskal-Wallis H = {stat:}, p-valor = {p:.9f}')
```

```python
df['group'] = df['Year'].astype(str) + '-' + df['Sector_Type']
posthoc = sp.posthoc_dunn(df, val_col='Total installed power (kW)', group_col='group', p_adjust='bonferroni')
print(posthoc)
```

Diferenças claras entre setores no mesmo ano:

2023 Industrial vs 2023 Residential → diferença significativa

2024 Industrial vs 2024 Residential → diferença significativa

Diferenças entre anos dentro de um setor:

Residential 2023 vs 2024 → diferença significativa

Industrial 2023 vs 2024 → não significativa (p = 1.0)

Interpretação prática para a RQ2

Residential UPACs cresceram de 2023 para 2024 (diferença significativa).

Industrial UPACs não tiveram crescimento estatisticamente significativo.

Industrial sempre com maior capacidade instalada que Residential (diferenças significativas entre setores).

# ANOVA

```python
# Filter districts
districts = ['Aveiro','Évora','Vila Real','Faro']
df_sample = df[df['District'].isin(districts)].copy()

# Map Voltage to Type
df_sample['Type'] = df_sample['Voltage level'].map({
    'AT':'Industrial',
    'MT':'Industrial',
    'BTN':'Residential',
    'BTE':'Residential'
})
df_sample = df_sample[df_sample['Type'].isin(['Residential','Industrial'])]

# Random 10% sample for computational efficiency
df_sample = df_sample.sample(frac=0.1, random_state=27)

# Extract Year and Quarter
df_sample['Year'] = df_sample['Quarter'].str[:4].astype(int)
df_sample['Q'] = df_sample['Quarter'].str[5:]  # e.g., Q1, Q2

# Sort data
df_sample.sort_values(['District','Type','Municipality','Quarter'], inplace=True)

# Compute previous quarter installed power
df_sample['Prev_power'] = df_sample.groupby(['District','Type','Municipality'])['Total installed power (kW)'].shift(1)

df_sample['Growth'] = (df_sample['Total installed power (kW)'] - df_sample['Prev_power']) / df_sample['Prev_power'] * 100
df_growth = df_sample.dropna(subset=['Growth'])

stat_levene, p_levene = levene(
    df_growth[df_growth['Type']=='Residential']['Growth'],
    df_growth[df_growth['Type']=='Industrial']['Growth']
)
print(f"Levene Test: W={stat_levene:.3f}, p={p_levene:.5f}")

model = ols('Growth ~ C(Type) * C(District)', data=df_growth).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nTwo-way ANOVA (Type x District) results:\n")
print(anova_table)

plt.figure(figsize=(12,6))
sns.barplot(data=df_growth, x='District', y='Growth', hue='Type', ci='sd', palette='Set2')
plt.title("Quarter-over-Quarter Installed Power Growth (%) by District and UPAC Type")
plt.ylabel("Growth (%)")
plt.xlabel("District")
plt.show()
```

# Random samples

```python
sample_sizes = [9000, 12500, 15000]
n_iterations = 10

results = []

districts = ['Aveiro','Évora','Vila Real','Faro']
df_sample = df[df['District'].isin(districts)].copy()

# Map Voltage to Type
df_sample['Type'] = df_sample['Voltage level'].map({
    'AT':'Industrial',
    'MT':'Industrial',
    'BTN':'Residential',
    'BTE':'Residential'
})

df_sample = df_sample[df_sample['Type'].isin(['Residential','Industrial'])]

# Extract Year and Quarter
df_sample['Year'] = df_sample['Quarter'].str[:4].astype(int)
df_sample['Q'] = df_sample['Quarter'].str[5:]  # e.g., Q1, Q2

# Sort data
df_sample.sort_values(['District','Type','Municipality','Quarter'], inplace=True)

# Compute previous quarter installed power
df_sample['Prev_power'] = df_sample.groupby(['District','Type','Municipality'])['Total installed power (kW)'].shift(1)
df_sample['Growth'] = (df_sample['Total installed power (kW)'] - df_sample['Prev_power']) / df_sample['Prev_power'] * 100
df_growth = df_sample.dropna(subset=['Growth'])

for size in sample_sizes:
    for i in range(1, n_iterations + 1):
        df_iter = df_growth.sample(n=size, replace=False, random_state=i+10)
        model = ols('Growth ~ C(Type) * C(District)', data=df_iter).fit()
        anova_iter = sm.stats.anova_lm(model, typ=2)

        # Append F and p-values for Type, District, Interaction
        results.append({
            'SampleSize': size,
            'Iteration': i,
            'F_Type': anova_iter.loc['C(Type)','F'],
            'p_Type': anova_iter.loc['C(Type)','PR(>F)'],
            'F_District': anova_iter.loc['C(District)','F'],
            'p_District': anova_iter.loc['C(District)','PR(>F)'],
            'F_Interaction': anova_iter.loc['C(Type):C(District)','F'],
            'p_Interaction': anova_iter.loc['C(Type):C(District)','PR(>F)']
        })

df_results = pd.DataFrame(results)

pd.set_option('display.float_format', '{:.4f}'.format)
print(df_results)

summary = df_results.groupby('SampleSize').agg(['mean','std'])
print("\nSummary of ANOVA metrics by sample size:\n", summary)
```

```python
df_pval_long = df_results.melt(
    id_vars=['SampleSize','Iteration'],
    value_vars=['p_Type','p_District','p_Interaction'],
    var_name='Factor',
    value_name='p_value'
)
df_pval_long['Factor'] = df_pval_long['Factor'].replace({
    'p_Type':'Type',
    'p_District':'District',
    'p_Interaction':'Type x District'
})

heatmap_data = df_pval_long.pivot_table(index='Iteration', columns=['SampleSize','Factor'], values='p_value')

plt.figure(figsize=(14,6))
sns.heatmap(heatmap_data, annot=True, fmt=".3f", cmap='coolwarm', cbar_kws={'label':'p-value'})
plt.title("ANOVA p-values across Iterations and Sample Sizes")
plt.ylabel("Iteration")
plt.show()
```

## **Analysis of Installed Capacity Growth between 2023 and 2024**

### **1. Research Context and Objective**

The goal of this analysis remains:

**Compare the evolution of installed capacity between 2023 and 2024 across Residential and Industrial UPACs to assess differences in growth patterns.**

We analyzed **quarterly growth** in installed power (kW) for Residential and Industrial UPACs across four districts: **Aveiro, Évora, Vila Real, and Faro**.

---

### **2. Why We Use ANOVA**

We used **two-way ANOVA** for these reasons:

1. **Two categorical independent variables**:

   * `Type` (Residential vs Industrial)
   * `District` (Aveiro, Évora, Vila Real, Faro)
   * Their **interaction** (`Type × District`) to assess whether growth differences between UPAC types vary by district.

2. **Continuous dependent variable**: quarterly growth in installed power.

3. **Efficiency**: ANOVA simultaneously tests main effects and interaction effects, avoiding multiple t-tests and inflated Type I error.

4. **Robustness check**: By performing **10 random sampling iterations** at multiple sample sizes (700, 1000, 1250), we can verify the **stability of the results** and ensure findings are not driven by outliers or specific sample draws.

---

### **3. Methodology**

1. **Data Preparation:**

   * Filtered four districts of interest.
   * Classified UPACs as **Residential** (BTN, BTE) or **Industrial** (AT, MT).
   * Randomly sampled **10% of records** to manage dataset size.
   * Removed zero installed power.
   * Extracted `Year` and `Quarter` from `Quarter` column.

2. **Growth Computation:**

   * Calculated **quarter-over-quarter growth** for each UPAC within each district:

$$
Growth = \frac{InstalledPower_q - InstalledPower_{q-1}}{InstalledPower_{q-1}} \times 100
$$


3. **Assumption Checks:**

   * **Independence:** ensured via random sampling.
   * **Normality:** ANOVA is robust with >30 observations per group.
   * **Homoscedasticity:** Levene’s test `W=11.886, p=0.00058` indicates unequal variances; results should be interpreted with **caution**, but ANOVA is fairly robust to moderate violations.

4. **Random Sampling Robustness:**

   * Ran **10 iterations** for sample sizes **700, 1000, 1250**.
   * Recorded F- and p-values for `Type`, `District`, and `Type × District` interaction.
   * Observed **consistent patterns of significance** across iterations and sample sizes.

---

### **4. Hypotheses**

**Effect of Type (Residential vs Industrial):**

* H₀: Mean growth is equal for Residential and Industrial UPACs.
* H₁: Mean growth differs between Residential and Industrial UPACs.

**Effect of District:**

* H₀: Mean growth is equal across all districts.
* H₁: Mean growth differs across districts.

**Interaction (Type × District):**

* H₀: The difference in growth between Residential and Industrial UPACs is consistent across districts.
* H₁: The difference in growth between UPAC types varies by district.

---

### **5. ANOVA Results**

**Original 10% sample:**

| Factor          | F     | p-value | Interpretation                                                |
| --------------- | ----- | ------- | ------------------------------------------------------------- |
| Type            | 11.60 | 0.00068 | Significant → Residential and Industrial UPAC growth differ   |
| District        | 4.46  | 0.0040  | Significant → Some districts grow faster than others          |

| ----------: | ----------: | ----------: | --------------: | --------------: | -----------------: | -----------------: |
|        9000 |        5.50 |     4.90e-0 |            8.00 |          0.0002 |               3.94 |             0.0166 |

**Observations:**

* `Type` is consistently significant across **all iterations and sample sizes** → Industrial and Residential growth differ.
* `District` is marginally to significantly significant depending on sample size → district-level differences exist.
* `Type × District` interaction becomes **significant for larger sample sizes**, confirming that the growth difference between UPAC types is district-specific.

---

### **6. Interpretation and Conclusions**

1. **Residential vs Industrial growth:**

   * Industrial UPACs grow **faster than Residential** on average.
   * Effect is consistent across random iterations → robust conclusion.

2. **District effect:**

   * Some districts, e.g., Aveiro and Faro, may experience higher growth than others.
   * Significance strengthens with **larger samples**, suggesting the effect is genuine.

3. **Interaction effect:**

   * Growth differences between Residential and Industrial UPACs **depend on district**.
   * E.g., Industrial may lead in Aveiro, while Residential is similar in Évora.

4. **Robustness via random sampling:**

   * Iterating across multiple samples and sizes confirms findings are **not an artifact of one sample**.
   * Larger samples stabilize F-values and p-values, giving more confidence in results.

**Overall conclusion:**
The evolution of installed capacity from 2023 to 2024 **differs by UPAC type** and these differences are **district-dependent**, highlighting that **policy, planning, and investment decisions should consider both UPAC type and local context**.
