```python
df_pop = pd.read_csv("1.csv")
df_pop
```

Renaming columns

```python
df_pop = df_pop.rename(columns={"01. Ano": "Ano", "03. Nome Região (Portugal)": "Nome_Reg", "10. Valor": "Num_pop" })
```

Targeting years of interest

```python
df_pop = df_pop[df_pop["Ano"].isin([2023, 2024])]
df_pop
```

Only keeping columns were "Filtro 1" (Sex) and "Filtro 2"(Age) are Total

```python
df_pop = df_pop[
    (df_pop["05. Filtro 1"].str.strip().str.lower() == "total") &
    (df_pop["06. Filtro 2"].str.strip().str.lower() == "total")
]

df_pop
```

Removing entries from countries that are not Portugal, this was done by removing rows where Nome_reg is NaN

```python
df_pop = df_pop.dropna(subset=["Nome_Reg"])	
```

Removing columns with all entries NaN

```python
df_pop = df_pop.dropna(axis=1, how='all')
```

Checking that all unique values from "Nome_Reg" have two entries, which means having an entry in 2023 and 2024

```python
contagem = df_pop["Nome_Reg"].value_counts()

print(contagem)

with open("contagem_regioes.txt", "w", encoding="utf-8") as f:
    for nome, freq in contagem.items():
        f.write(f"{nome}\t{freq}\n")
```

```python
duplicadas = df_pop[df_pop.duplicated()]
print(f"Número de linhas duplicadas: {len(duplicadas)}")
```

In order to join the previous datasets a 3rd one was needed to connect district to region to population per district

Link: https://public.opendatasoft.com/explore/assets/georef-portugal-concelho/export

```python
df_md = pd.read_csv("georef-portugal-concelho.csv", sep=';', encoding='utf-8')
df_md
```

Removing not so important columns

```python
colunas_para_remover = [
    "Geo Point", 
    "Geo Shape", 
    "Year", 
    "Iso 3166-3 Area Code"
]

df_md = df_md.drop(columns=colunas_para_remover)
df_md
```

Normalise column Nome_Reg (remove extra spaces and lowercase letters) for proper merge later

Select only columns that are important for the merge from df_md


```python
df_pop['Nome_Reg_clean'] = df_pop['Nome_Reg'].str.strip().str.lower()

df_md_subset = df_md[['Official Name Municipality lower case', 'Official Name District']]
```

Merge of the datasets and then removing the "extra" columns

```python
df_merged = df_pop.merge(
    df_md_subset,
    left_on='Nome_Reg_clean',
    right_on='Official Name Municipality lower case',
    how='left'
)

df_merged = df_merged.drop(columns=['Nome_Reg_clean', 'Official Name Municipality lower case'])

df_merged
```

```python
duplicadas = df_merged[df_merged.duplicated()]
print(f"Número de linhas duplicadas: {len(duplicadas)}")
```

Removing not so important columns

```python
colunas_para_remover = [
    "05. Filtro 1", 
    "06. Filtro 2", 
    "08. Escala ", 
    "04. Âmbito Geográfico"
]

df_merged = df_merged.drop(columns=colunas_para_remover)
df_merged
```

Understanding where there are null values.

Since there are from either Madeira and Azores, they can be removed as the original dataset does not have information for those regions. Portugal as a whole can be removed too

```python
linhas_com_null = df_merged[df_merged.iloc[:, -1].isna()]

linhas_com_null
```

```python
df_merged = df_merged.dropna(subset=[df_merged.columns[-1]])
df_merged
```

Creating a new column that has all the population from the district that the value from the column nome_reg belongs too.

```python
df_merged = df_merged.copy()

df_merged["pop_distrito"] = (
    df_merged.groupby(["Ano", "Official Name District"])["Num_pop"]
    .transform("sum")
)
df_merged
```

ensure that names are consistent, that data type is the same and then convert districts to lowercase (to avoid failures due to upper/lowercase letters) to prepare the dataframes to merge

```python
df_merged = df_merged.rename(columns={"Official Name District": "distrito"})

df["Ano"] = df["Ano"].astype(int)
df_merged["Ano"] = df_merged["Ano"].astype(int)

df["distrito"] = df["distrito"].str.lower().str.strip()
df_merged["distrito"] = df_merged["distrito"].str.lower().str.strip()
```

```python

df_final = pd.merge(
    df,
    df_merged[["Ano", "distrito", "pop_distrito"]],
    on=["Ano", "distrito"],
    how="left"
)
```

```python
df_final
```

```python
duplicadas = df_final[df_final.duplicated()]
print(f"Número de linhas duplicadas: {len(duplicadas)}")
```

```python
df_final = df_final.drop_duplicates()
df_final
```

Only keeping the target years (2023, 2024)

```python
df_final = df_final[df_final["Ano"].isin([2023, 2024])]
df_final
```
